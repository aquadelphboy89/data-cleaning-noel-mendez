## 7. Análisis univariado de variables numéricas
Realizar histogramas y boxplots para analizar la distribución de las variables numéricas.

La primera pregunta que nos podemos hacer es, ¿Las incidencias se notifican al momento o hay un retraso importante?

```python
# Histogram for a numerical variable
df['ReportingDelay'].hist()
plt.title('Delay between Offense and Call')
plt.xlabel('Delay (in seconds) between Offense and Call')
plt.ylabel('Number of Incidents')
plt.show()
```

Tras analizar la variable ReportingDelay, la cual mide en segundos el tiempo transcurrido entre la ocurrencia del delito y su reporte, el histograma muestra que la mayor parte de los registros se concentran entre los 30.000 y 70.000 segundos (aproximadamente entre 8 y 20 horas). Esto indica que la mayoría de los incidentes no se reportan de forma instantánea, sino que existe un intervalo de tiempo considerable, probablemente debido al proceso de descubrimiento del delito o al tiempo de respuesta administrativo. También hay que tener en cuenta que la hora del crimen se registra de forma genérica sin hora específica (se marca como 0:00) mientras que la hora del reporte sí se registra, siendo esta la hora real. Como conclusión de estos datos, podemos sacar que casi todos los líos se avisan dentro del mismo día, antes de que pasen 24 horas, lo que significa que el sistema funciona rápido y no se amontonan papeles de días anteriores. También se ve que es muy raro que un aviso sea inmediato o al segundo de pasar el problema, seguramente por incluir el incidente a las 00:00 tal y como comentábamos. Finalmente, el gráfico nos dice que los registros son bastante ordenados, ya que no aparecen casos perdidos que tarden muchísimos días en aparecer en la lista, lo que da confianza en cómo se guardan los datos.

```python

# Boxplot
df.boxplot(column='ReportingDelay')
plt.title('Reporting Delay Distribution')
plt.ylabel('Delay (in seconds) between Offense and Call')
plt.show()
```

Si nos fijamos ahora en el boxplot para el tiempo de retraso, vemos que los datos están muy ordenados y confirman lo que ya sospechábamos. La parte central de la caja, que es donde se amontonan la mitad de todos los casos, se queda entre los 30.000 y 70.000 segundos, lo que nos dice que el tiempo de aviso es bastante previsible y no es un caos. La raya verde del medio está bien puesta, lo que indica que casi todos los avisos siguen el mismo ritmo cada día, cerca de los 50.000 segundos, confirmando que la mayoría de los registros tienen un retraso de muchas horas. Además, en este gráfico no vemos puntos sueltos por arriba, lo que significa que prácticamente todos los incidentes se quedan registrados dentro de las primeras 24 horas sin que haya casos que se olviden o tarden muchísimos días en aparecer. Al final, la posición de esta caja nos vuelve a demostrar que el reloj del sistema manda mucho, ya que la hora del incidente queda reflejada como las 0:00, la caja sube hacia arriba y no deja que veamos avisos que empiecen desde el cero.

La siguiente pregunta que nos podríamos hacer es ¿a qué hora del día se reportan más incidentes? Esto es útil saberlo para saber a que hora del día se requieren de más profesionales para asi poder atender con mayor rapidez.

```python
df['Call_Hour'].hist()
plt.title('Distribution of Incidents by 1-Hour Intervals')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Incidents')
plt.xticks(range(0, 25))
plt.show()
```

Al analizar la distribución de los incidentes por franjas de una hora, se observa un patrón de actividad con picos muy localizados que reflejan la dinámica de la ciudad. El gráfico muestra un descenso de actividad muy marcado durante las primeras horas de la madrugada, alcanzando su punto más bajo entre las 02:00 y las 07:00 horas, lo cual es coherente con el ciclo de descanso general. A partir de las 07:00 se produce un repunte significativo que se mantiene hasta las 09:00, coincidiendo con el inicio de la actividad urbana y los desplazamientos matutinos. Sin embargo, el pico máximo de incidentes se sitúa claramente en la franja de la tarde, concretamente entre las 14:00 y las 16:00 horas, siendo el momento del día con mayor volumen de registros. Es destacable que, tras una ligera disminución, la actividad vuelve a intensificarse notablemente al final del día, especialmente en la franja de 21:00 a 23:00 horas. Este comportamiento sugiere que la demanda de servicios de emergencia en San Francisco tiene tres momentos críticos: la entrada al trabajo, el bloque central de la tarde y el periodo de ocio nocturno. Esta variable horaria se presenta como un indicador altamente fiable para entender la carga de trabajo real, ya que permite identificar con precisión estos periodos de máxima afluencia.

```python

# Boxplot
df.boxplot(column='Call_Hour')
plt.title('Call Hour Distribution')
plt.ylabel('Hour of the Day')
plt.show()
```

Con esto confirmamos la robustez de los datos y la ausencia de valores atípicos, distribuyendo la actividad de manera equilibrada a lo largo de las 24 horas del día. Se observa que la mediana se sitúa en torno a las 14:00 horas, lo que coincide con el inicio del pico máximo de actividad identificado en el histograma. El rango intercuartílico, representado por la caja, abarca aproximadamente desde las 09:00 hasta las 18:00 horas, lo que indica que el 50% central de todos los incidentes ocurre durante la jornada laboral y la tarde, subrayando la alta densidad de casos en estas franjas. Por otro lado, los "bigotes" del gráfico se extienden hasta cubrir la totalidad del espectro horario desde las 00:00 hasta las 23:00, lo que demuestra que, aunque existen periodos de menor intensidad como la madrugada, la actividad policial es constante y no existen horas del día con comportamientos extraños o fuera de lo común. Esta visualización refuerza la idea de que la distribución horaria es representativa del ritmo de una ciudad activa, con una dispersión de datos previsible y bien estructurada.

## 8. Análisis univariado de variables categóricas
Realizar gráficos de barras para analizar la frecuencia de las variables categóricas.

Esta sección es más interesante ya que tenemos más columnas categoricas. La primera pregunta que nos podemos plantear es, ¿Qué dia de la semana se produce más crimenes/incidentes? Y, ¿Qué dia se producen menos? Veamoslo.

```python
counts = df['Offense_DayName'].value_counts()
ax = counts.plot(kind='bar')

ax.bar_label(ax.containers[0]) # Añade etiquetas a cada barra para hacerlo más visual

plt.title('Crimes by Day of the Week')

plt.xlabel('Day')

plt.ylabel('Number of Crimes')

plt.show()
```

Al observar el gráfico de barras, se identifica que la mayor concentración de incidentes ocurre entre el viernes y el lunes. El martes también se registra una actividad considerable, aunque representa aproximadamente la mitad de lo observado en los días pico.

Finalmente cal destacar que el miércoles y jueves presentan cifras mínimas. Dado que este dataset registra datos de incidencias entre un periodo corto de tiempo, esta caída podría deberse a la ventana de tiempo de la extracción de datos y no necesariamente a una ausencia real de criminalidad en esos días de la semana. Para ello, para afirmar que realmente esos dias no hay una actividad criminal, se deberia ampliar el dataset con una mayor frecuencia de tiempo.

La siguiente pregunta que nos podemos plantear es ¿Qué tipo de crimenes son los más usuales en San Francisco?

```python
# Cogemos solo los 10 tipos de crimenes con mayor frecuencia para que el gráfico sea legible, ya que hay muchos tipos de crimen diferentes
counts = df['OriginalCrimeTypeName'].value_counts().head(10) 

ax = counts.plot(kind='bar')

ax.bar_label(ax.containers[0]) # Añade etiquetas a cada barra para hacerlo más visual

plt.title('Crimes by Crime Type')

plt.xlabel('Crime name')

plt.ylabel('Number of Crimes')

plt.show()
```

De esta forma, podemos decir que los incidentes más usuales son 'Traffic Stop' (Controles de Tráfico) y 'Passing Call' (Rondas de vigilancia preventiva), ambos con más de 1000 casos. Esto indica que el dataset está compuesto principalmente por labores de prevención y patrullaje más que por reportes de criminalidad violenta. El primer incidente de naturaleza delictiva o de queja ciudadana que aparece es 'Suspicious Person' (Persona sospechosa), con una frecuencia notablemente menor (menos de 700 casos).

La siguiente pregunta que nos podríamos hacer es ¿Cómo suelen resolverse los  crímnes?

```python
counts = df['Disposition'].value_counts()
ax = counts.plot(kind='bar')

ax.bar_label(ax.containers[0]) # Añade etiquetas a cada barra para hacerlo más visual

plt.title('Crimes by Disposition')

plt.xlabel('Disposition')

plt.ylabel('Number of Crimes')

plt.show()
```

Al analizar el gráfico resultante, destaca que la mayoría de los casos se cierran bajo el código HAN (Handled) con 2819 registros, seguido por CIT (Citation) con 1423.

Con esta distribución podemos intuir que la actividad policial en este dataset es mayoritariamente de baja intensidad ya que la mayoría de los incidentes se resuelven mediante la gestión directa del oficial en el lugar (HAN) o mediante multas (CIT). Por el contrario, las resoluciones que implican medidas más severas como el arresto (ARR) o situaciones donde no se encuentra el incidente (NCR) representan una parte mínima de la muestra, lo que sugiere una alta eficiencia en la respuesta o que las llamadas son por motivos muy específicos y localizados.

La ultima pregunta, podría ser ¿Dónde suelen ocurrir más delitos?

```python
counts = df['AddressType'].value_counts()
ax = counts.plot(kind='bar')

ax.bar_label(ax.containers[0]) # Añade etiquetas a cada barra para hacerlo más visual

plt.title('Crimes by Address Type')

plt.xlabel('AddressType')

plt.ylabel('Number of Crimes')

plt.show()
```

Con esto, podemos observar una clara predominancia de los incidentes ocurridos en direcciones específicas o inmuebles (Premise Address), con 5858 casos, lo que sitúa el foco de la actividad policial principalmente en viviendas, locales comerciales o edificios. Aunque su frecuencia es mínima, su aparición en el gráfico sugiere la necesidad de un estudio posterior para determinar su significado técnico, pudiendo representar un código interno para direcciones fuera de jurisdicción o ubicaciones genéricas del distrito que el sistema no ha podido categorizar de forma automática.

## 9. Análisis bivariado entre variables
Explorar relaciones entre variables numéricas y categóricas mediante scatterplots y tablas cruzadas.

Lo primero que podemos cruzar es el mes que ocurren los incidentes frente al mes que se reportaron.

```python
# Scatterplot
plt.scatter(df['Offense_Month'], df['Call_Month'])
plt.title('Offense Month vs Call Month')
plt.xlabel('Offense Month')
plt.ylabel('Call Month')
plt.show()
```

Este gráfico es poco ilustrativo e interesante. Al tener que las notificaciones no tardan ni 24 horas es normal, que salgan solo dos puntos, los de coordenadas (3,3) y (4,4) ya que los incidentes de cada mes se notifican ese mismo dia, en otras palabras, ese mismo mes.

La siguiente es comparar la hora de la llamada (sin los minutos) frente a la diferencia de tiempo en segundos desde que ocurrio el incidente hasta que se notifico. 

```python
plt.scatter(df['Call_Hour'], df['ReportingDelay'])
plt.title('Call Hour vs Reporting Delay')
plt.xlabel('Call_Hour')
plt.ylabel('ReportingDelay')
plt.show()
```

Con esto, podemos concluir que los avisos que se hacen por la noche tardan más en registrarse que los de la mañana.

Veamos ahora podemos cruzar los dias que se cometieron los delitos frente los dias que se registraron los delitos.

```python
# Crosstab for two categorical variables   
pd.crosstab(df['Offense_DayName'], df['Call_DayName'])
```

De nuevo, como no pasan ni 24 horas desde que se cometió el delito hasta que se notifico, es logico que todos los valores excepto la diagonal sean 0 ya que se notifica el mismo dia que se comete el delito.

Crucemos ahora el tipo de crimenes frente al dia de la semana, para corrobar que tipo de delitos son los más reportados cada dia de la semana.

```python
# Como hay muchos tipos de crimen diferentes, vamos a quedarnos solo con los 10 más frecuentes.
top_10_crimenes = df['OriginalCrimeTypeName'].value_counts().nlargest(10).index
df_top = df[df['OriginalCrimeTypeName'].isin(top_10_crimenes)]
```

```python
pd.crosstab(df_top['OriginalCrimeTypeName'], df_top['Call_DayName'])
```

Teniendo en cuenta que hemos filtrado los crimenes a los 10 más frecuentes, podemos ver que de viernes a lunes, así como los martes, la actividad predominante del reporte corresponde a Traffic Stop (controles de tráfico) y Passing Call (rondas de vigilancia). Por su parte, los jueves destacan por un reporte de Homeless Complaint (quejas por indigencia). Esto no significa especialmente que dichos dias, sean los crímenes con mayor numero de reportes ya que podria ser que uno o varios de los días tenga un mayor número de reportes que no es muy tipico de forma génerica pero si de ese dia. Tambien es relevante notar la ausencia de reportes los miércoles en este gráfico; esto es consistente con el análisis univariado previo, donde se detectó que dicha jornada apenas hubo un reporte, el cual, al no pertenecer a este 'Top 10', queda excluido de esta visualización bivariada.

Podemos hacer lo mismo, pero en vez de cruzarlo con el reporte del crimen, cruzarlo por el delito ocurrido, aunque el resultado sea identico ya que tal y como vimos, los delitos sucedidos se notifican el mismo día. Comprobemoslo.

Efectivamente, la tabla es identica. Las conclusiones son las mismas.

## 10. Visualización de correlaciones
Calcular y visualizar la matriz de correlación entre variables numéricas usando un heatmap.

```python
# Correlation matrix
cols_numericas = ['ReportingDelay', 'Call_Hour', 'Call_Month', 'Offense_Month'] 
corr = df[cols_numericas].corr() # Para la correlación solo tiene sentido coger las variables numéricas.
# No incluimos los identificadores como CrimeId y AgencyId ya que no aportan información relevante para el análisis.
sns.heatmap(corr, annot=True)
plt.show()
```

El análisis de la matriz de correlación muestra que existe una relación perfecta (1.0) entre el mes de la llamada (Call_Month) y el mes del incidente (Offense_Month), lo que indica que en la totalidad de los registros analizados el reporte se produce en el mismo mes en que ocurre el evento. Esto es lógico porque ya hemos visto el reporte se produce en el mismo dia que ocurre el evento. Por otro lado, destaca una correlación extremadamente alta, es decir una relación perfecta de 1.0, entre la hora de la llamada (Call_Hour) y el retraso reportado (ReportingDelay); esto revela que, en este dataset específico, el valor de "retraso" no mide el tiempo transcurrido desde el crimen, sino que representa simplemente la hora del día convertida a segundos desde la medianoche. Finalmente, se observa que no existe una correlación significativa entre el mes y la hora de los incidentes, lo que sugiere que la distribución horaria de los delitos se mantiene constante y es independiente de la época del año.
